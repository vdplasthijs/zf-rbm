```python
## Load data

time_cutoff_end = 4000 # hard coded for now
# time_cutoffs = {''}
recordings = {}
for data_set, data_path in data_sets.items():
    recordings[data_set] = Zecording(path=base_path + data_path, kwargs={'ignorelags': True,
                                              'forceinterpolation': False,
                                              'ignoreunknowndata': True,# 'parent': self,
                                              'loadram': True})  # load data
rec = recordings['20180912-Run01']

print(rec)

## Select regions for evaluation (via zbrain atlas)
# 'wb' = 'whole-brain'; selected by selecting all zbrian regions (0:294)
# This selection is used throughout - neurons that are not labelled are not considered (~5%) to exclude artefacts
regions = {'wb': np.arange(294)}  # 'rh1': np.array([218]), 'rhall': np.array([113]),
inds_wb = np.unique(scipy.sparse.find(rec.labels[:, np.arange(294)])[0])
inds_all_dict = {'wb': inds_wb}
selected_neurons = {}
n_sel_cells = {}
train_data = {}
test_data = {}
full_data = {}
train_times = rec.times[:time_cutoff_end]
test_times = rec.times[time_cutoff_end:]

for ir in list(regions.keys()):
    selected_neurons[ir] = np.unique(scipy.sparse.find(rec.labels[:, regions[ir]])[0])
    n_sel_cells[ir] = len(selected_neurons[ir])
    train_data[ir] = rec.spikes[selected_neurons[ir], :time_cutoff_end]
    test_data[ir] = rec.spikes[selected_neurons[ir], time_cutoff_end:]
    full_data[ir] = rec.spikes[selected_neurons[ir], :]
```

```python
## Load RBMs

## Thijs 17Nov2019: apologies this section is very lengthy currently.. Will clean up in the future.
# Currently it only selects rbm 2019-05-23-1921

all_rbm_names = [x for x in os.listdir(multiple_rbm_path) if x[-5:] == '.data'] #  and x.split('_')[1] == '2019-03-26(Run09)'
# all_rbm_names.append('RBM7ADAM_20180912-Run01_wb_train0-4000_M100_l1-1e-02_nit100000_nmc10_duration92635s_timestamp2019-05-23-1921.data')
all_generated_names = os.listdir(multiple_rbm_path + 'generated/')
gen_links = {}
for gen_name in all_generated_names:
    ts = [x.lstrip('timestamp') for x in gen_name.split('_') if x[:9] == 'timestamp']
    if len(ts) == 1:
        gen_links[ts[0]] = multiple_rbm_path + 'generated/' + gen_name
    
n_rbms = len(all_rbm_names)
float_features = ['nhu', 'n_cells', 'l1', 'nit', 'duration', 'nmc', 'lr', 'lr_final'] + ['means_v', 'means_h', 'pwcorr_vh', 'pwcorr_vv', 'pwcorr_hh']
str_features = ['timestamp', 'dataset', 'region', 'dataselection', 
                 'hiddenprior', 'fullpath', 'rbmtype'] 
                
df_lookup = pd.DataFrame({**{x: np.zeros(n_rbms) for x in float_features}, 
                          **{x: '0' for x in str_features}})

all_rbm_dict = {} # save all rbms in memory to create mappings
mapping = {}
inverse_mapping = {}
low_dyn_train = {}
low_dyn_test = {}
low_dyn_full = {}

for irbm, rbm_name in enumerate(all_rbm_names):
    for iy, y in enumerate(rbm_name.split('_')):
        ts_found = False
        if 'timestamp' in y:
            ts_rbm = y.lstrip('timestamp').rstrip('.data')
            ts_found = True
    if not ts_found:
        print(f'ts not found for {rbm_name}, aborting')
        continue
        
    if ts_rbm != '2019-05-23-1921':
        continue
    else:
        print('ts found!')
        
    ## Create look up table
    df_lookup['fullpath'].iat[irbm] = multiple_rbm_path + rbm_name
    tmp_rbm = pickle.load(open(multiple_rbm_path + rbm_name, 'rb'))
    
    try:
        tmp_rbm.track_mm
        add_track_mm = True
    except AttributeError:  # not all RBMs have tracked moments -> exclude these
        add_track_mm = False
        if ts_rbm != '2019-05-23-1921':
            continue
        
    all_rbm_dict[ts_rbm] = tmp_rbm
    mapping[ts_rbm] = lambda high_data, t=ts_rbm: np.transpose(all_rbm_dict[t].mean_hiddens(high_data.transpose()))
    inverse_mapping[ts_rbm] = lambda low_data, t=ts_rbm: np.transpose(all_rbm_dict[t].mean_visibles(low_data.transpose()))
    
    df_lookup['l1'].iat[irbm] = tmp_rbm.l1
    df_lookup['lr'].iat[irbm] = tmp_rbm.learning_rate_init
    df_lookup['lr_final'].iat[irbm] = tmp_rbm.lr_final
    df_lookup['n_cells'].iat[irbm] = tmp_rbm.n_v
    df_lookup['nhu'].iat[irbm] = tmp_rbm.n_h
    df_lookup['nit'].iat[irbm] = tmp_rbm.n_iter  #TODO: change to number of updates?
    df_lookup['nmc'].iat[irbm] = tmp_rbm.N_MC
    df_lookup['hiddenprior'].iat[irbm] = tmp_rbm.hlayer.nature
    
    if add_track_mm:
        for mom in ['means_v', 'means_h', 'pwcorr_vh', 'pwcorr_vv', 'pwcorr_hh']:
            try:
                df_lookup[mom].iat[irbm] = tmp_rbm.track_mm[mom][-1]
            except KeyError:  # some track_mm-RBMs lack the pwcorr_vv 
                df_lookup[mom].iat[irbm] = None
        
    for iy, y in enumerate(rbm_name.split('_')):
        if iy == len(rbm_name.split('_')) - 1:
            y = y.rstrip('.data')
        if iy == 0 and y[:3] == 'RBM': 
            df_lookup['rbmtype'].iat[irbm] = y # first one is rbm type
        elif iy == 1 and y[:3] == '201':
            df_lookup['dataset'].iat[irbm] = y  # second one is data name
        elif y in list(regions.keys()): #  iy == 2 and
            df_lookup['region'].iat[irbm] = y  # third one is region spec
            tmp_region = y
        elif iy == 3 and y[:5] == 'train':
            df_lookup['dataselection'].iat[irbm] = y
        elif 'duration' in y:
            df_lookup['duration'].iat[irbm] = y.lstrip('duration').rstrip('s')
        elif 'timestamp' in y:
            df_lookup['timestamp'].iat[irbm] = y.lstrip('timestamp')
      
    if test_data[tmp_region].shape[0] == all_rbm_dict[ts_rbm].n_v:
        # Create low dim dynamics
        low_dyn_test[ts_rbm] = mapping[ts_rbm](test_data[tmp_region])
        low_dyn_train[ts_rbm] = mapping[ts_rbm](train_data[tmp_region])
        low_dyn_full[ts_rbm] = mapping[ts_rbm](full_data[tmp_region])
    
df_lookup = df_lookup[df_lookup['timestamp']  != '0'] 
df_lookup = df_lookup[df_lookup.dataset.str.contains('20180912')]

assert len(np.unique(df_lookup['timestamp'])) == len(df_lookup['timestamp'])
df_lookup['n_dims'] = df_lookup['nhu'] * df_lookup['n_cells']

print(f'Number of low dyns: {len(low_dyn_full)}')
```

# Compute mesoscopic/regional functional connectivity matrix

```python
## Compute mesoscopic (i.e. regional) connectivity matrix

def meso_conn_metric(matrix, method='threshold', threshold=0.1):
    """Metric to summarise all elements of one matrix"""
    
    if method == 'mean':
        return np.mean(np.abs(matrix))
    elif method == 'threshold':
        return (np.sum(matrix >= threshold) / matrix.size)

## Loop through all region-region combinations and compute functional connectivity
method_funct = 'mean'  # choose method to 

n_regions = len(regions)
meso_connectivity = np.zeros((n_regions, n_regions))  # matrix of mesoscopic connectivity
for ir1, r1 in tqdm(enumerate(regions)):  # double loop through regions
    for ir2, r2 in enumerate(regions):
        inds_r1 = scipy.sparse.find(grid_labels[:, r1])[0] # neuron indices of both regions
        inds_r2 = scipy.sparse.find(grid_labels[:, r2])[0] # neuron indices of both regions
        if (len(inds_r1) > 0) and (len(inds_r2) > 0):
            inds_comb = scipy.sparse.find(grid_labels[:, np.array([r1, r2])])[0] # neuron indices of both regions
            labelled_weights = all_rbm_dict[ts].weights[:, inds_comb]  # select weights of this selection
            functional_connectivity = np.dot(labelled_weights.T, mean_var[:,np.newaxis] * labelled_weights) # compute functional connectivity matrix of subselection
            meso_connectivity[ir1, ir2] = meso_conn_metric(
                functional_connectivity, method=method_funct)  # compute mesoscopic connectivity
        else: # no neurons in one (or two ) of the regions:
            meso_connectivity[ir1, ir2] = 0  # default option to prevent nans
```

```python
# ## Select regions that include both left and right counterparts:
# ir = 0
# region_names_arr = np.array(region_names)
# bool_rl = np.zeros(n_regions, dtype='bool')
# while ir < n_regions:
#     r = region_names_arr[ir]
#     if r[-4:] == 'left':
#         if r[:-4] + 'right' in region_names:
#             bool_rl[ir] = True
#             bool_rl[np.where(region_names_arr == r[:-4] + 'right')[0][0]] = True
#     else:  # only consdier left for brevity
#         ir += 1
#         continue
#     ir += 1
# region_names_rl = list(region_names_arr[bool_rl])

# meso_connectivity_rl = meso_connectivity[bool_rl, :][:, bool_rl]
# struct_conn_select_rl = struct_conn_select[bool_rl, :][:, bool_rl]

## Sort matrix by hemisphere: 
def sort_rl_matrix(mat):
    assert mat.shape[0] % 2 == 0
    half_ind = int(0.5 * mat.shape[0])
    mat_sort = np.zeros_like(mat)
    mat_sort[:half_ind, :half_ind] = mat[::2, :][:, ::2]
    mat_sort[:half_ind, half_ind:] = mat[::2, :][:, 1::2]
    mat_sort[half_ind:, :half_ind] = mat[1::2, :][:, ::2]
    mat_sort[half_ind:, half_ind:] = mat[1::2, :][:, 1::2]
    return mat_sort
meso_connectivity_rl_sort = sort_rl_matrix(meso_connectivity)
struct_conn_select_rl_sort = sort_rl_matrix(struct_conn_select)
```
